```python
import spacy

from spacy.lang.en.stop_words import STOP_WORDS

len(STOP_WORDS)
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("I'm Jonaid Sharif and Junior QA Automation Engineer skilled in Python and JavaScript with hands-on experience in Selenium, WebdriverIO, JMeter, and Postman. I’ve built and executed end-to-end test frameworks for UI and API validation in real-world e-commerce projects. Strong in performance testing, defect tracking, and automation strategy.")

for token in doc:
     if token.is_stop:
          print(token)
```

```python
def preprocess(text):
  doc = nlp(text)

  no_stop_words = [token.text for token in doc if not token.is_stop]
  return no_stop_words
```

```python
def preprocess(text):
  doc = nlp(text)

  no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
  return no_stop_words
```
